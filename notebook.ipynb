```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean_height = 175.6  # average height in cm
std_dev_height = 7.6  # standard deviation in cm
num_samples = 10  # number of samples

# Generate random samples from the normal distribution
heights = np.random.normal(mean_height, std_dev_height, num_samples)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(heights, bins=80, density=True, alpha=0.6, color='grey', edgecolor='black')

# Adding a title and labels
plt.title('Histogram of Adult Male Heights in France')
plt.xlabel('Height (cm)')
plt.ylabel('Density')


plt.show()

# Calculate and print the average height
average_height = np.mean(heights)
print(f"The average height of the sampled adult males is: {average_height:.2f} cm")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean_height = 175.6  # average height in cm
std_dev_height = 7.6  # standard deviation in cm
num_samples = 100  # number of samples to take

# Size of each sample
sample_size = 100

# Generate sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.normal(mean_height, std_dev_height, sample_size)
    sample_means.append(np.mean(sample))

# Plotting the histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=80, density=True, alpha=0.6, color='grey', edgecolor='black')

# Adding a title and labels
plt.title('Sampling Distribution of Sample Means (Sample Size = 100)')
plt.xlabel('Sample Mean Height (cm)')
plt.ylabel('Density')

plt.axvline(mean_height, color='red', linestyle='dashed', linewidth=1, label='Population Mean')
plt.legend()

plt.show()

# Calculate and print the average of sample means
average_sample_mean = np.mean(sample_means)
print(f"The average of the sampled means is: {average_sample_mean:.2f} cm")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Outcomes of the fair coin
outcomes = ['Heads', 'Tails']
# Probabilities for each outcome
probabilities = [0.5, 0.5]

# Creating the bar plot for the PMF
plt.figure(figsize=(8, 5))
plt.bar(outcomes, probabilities, color=['skyblue', 'lightcoral'], edgecolor='black')

# Adding title and labels
plt.title('PMF of a Fair Coin')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.ylim(0, 1)

# Adding text labels on bars
for i, v in enumerate(probabilities):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
